```python
from diffusers import AutoPipelineForText2Image
import torch

pipeline = AutoPipelineForText2Image.from_pretrained(
	"kandinsky-community/kandinsky-2-2-decoder", torch_dtype=torch.float16
).to("cuda")
generator = torch.Generator("cuda").manual_seed(31)
```

```python
image = pipeline(prompt = "a meteor shower on a night sky, in antartica, detailed, only a few meteors",
                 negative_prompt="blur"
                ).images[0]
```

```python
image
```

```python
from PIL import Image
image.save('meteor.png')
```

```python
for i in range(0,10):
  image = pipeline(prompt = "a meteor shower on a night sky, in antarctica, detailed, clear sky",
                   negative_prompt="blur",
                   ).images[0]
  image.save(f'meteor{i}.png')
```
